**Control Flow in Python [if,elif,else]**

```python
#if,else,elif

x=8
if x>10:
    print("X is greater than 10")
elif x>5:
    print("X is greater than 5 but less than 10")
else:
    print("X is less than 5 or 5")
```

**Nested If Statement**

```python
#Nested Loop

x=10

if x>0:
    print("X is Positive number")
    if x%2==0:
        print("X is Even")
    else:
            print("X is Odd")
else:
    print("X is non-positive")
```
